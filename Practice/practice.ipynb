```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(temperature=0.6)
name = llm.invoke("I want to open a restuarant for Indian food. Suggest a good name for this.")
print(name)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template_name = PromptTemplate(
    input_variables=["cuisine"],
    template="I want to open a restuarant for {cuisine} food. Suggest a good name for this."
)

prompt_template_name.format(cuisine="Italian")
```

```python
#LLMchain

chain = prompt_template_name | llm
#chain.invoke("Mexican")
response = chain.invoke({"cuisine": "Mexican"})
print(response)
```

```python
#SimpleSequentialChain

llm = ChatOpenAI(temperature=0.6)

prompt_template_name = PromptTemplate(
    input_variables=["cuisine"],
    template="I want to open a restuarant for {cuisine} food. Suggest a good name for this."
)

name_chain = prompt_template_name | llm

prompt_template_items = PromptTemplate(
    input_variables=["restaurant_name"],
    template="Suggest some menu items for {restaurant_name}. Return the items as a comma-separated list."
)

food_items_chain = prompt_template_items | llm

sequential_chain = name_chain | food_items_chain

response = sequential_chain.invoke({"cuisine": "Mexican"})
print(response.content)
```
